```python
select_away_team = """SELECT * FROM away_team LIMIT 10"""
cursor = conn.cursor()
cursor.execute(select_away_team)
result = cursor.fetchall()

p = []

tbl_Away = "<tr><th>game_id</th><th>Opp_Abbrev</th><th>Opp_Score</th><th>Opp_Pace</th><th>Opp_efg_pct</th><th>Opp_off_rtg</th></tr>"
p.append(tbl_Away)

for row in result:
    a = "<tr><td>%s</td>"%row[0]
    p.append(a)
    b = "<td>%s</td>"%row[1]
    p.append(b)
    c = "<td>%s</td>"%row[2]
    p.append(c)
    d = "<td>%s</td>"%row[3]
    p.append(d)
    e = "<td>%s</td>"%row[4]
    p.append(e)
    f = "<td>%s</td>"%row[5]
    p.append(f)
    
p = (''.join(p))
print(p)
```
